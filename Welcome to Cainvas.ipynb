<h1 style="font-weight: bold">What is Cainvas?</h1>
<br>

Cainvas is an online development environment to put ML models into edge devices including mobile, raspberry pi and arduino.

### Students or AI reaserchers
It is built for extremely fast prototyping, data visualization and ML model training.

### Data Scientists
It lets you upload, visualize, train and compile ML models with Python/C++ notebooks easily shareable with the world.

### Embedded ML engineers
Bring your favourite device and choice of pretrained ML model, and make your device a smart device. With Cainvas DIY projects, you can make your home smart with applications like, smart switch listening to your voice command.

# **Data science**

With Cainvas you can harness the full power of popular Python libraries to analyze and visualize data. The code cell below uses **numpy** to generate some random data, and uses **matplotlib** to visualize it. To edit the code, just click the cell and start editing.

```python
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```

# **deepC**

**[deepC Compiler and inference framework](https://github.com/ai-techsystems/deepC)** is designed to **enable and perform** deep learning neural networks by focussing on features of small form-factor devices like micro-controllers, eFPGAs, cpus and other embedded devices like [raspberry-pi](https://www.raspberrypi.org/), [odroid](https://www.hardkernel.com/), [arduino](https://www.arduino.cc/), [SparkFun Edge](https://www.sparkfun.com/products/15170), [risc-V](https://www.amazon.com/Seeed-Studio-Sipeed-Maixduino-RISC-V/dp/B07SW9ZWQQ), mobile phones, x86 and arm laptops among others.

![edge Devices](https://raw.githubusercontent.com/ai-techsystems/deepC/master/misc/edgeDevices.png "supported edge devices")

deepC also offers ahead of time compiler producing optimized executable based on [LLVM compiler tool chain](https://llvm.org/) specialized for deep neural networks with [ONNX](https://onnx.ai/) as front end.

### **deepC works with Cainvas platform with full synchronization**

### Let's see deepC workflow below:

```python
import deepC.dnnc as dc

x=dc.arange(1256)/100
sinx=dc.sin(x)
cosx=dc.cos(x)

# deepC supports all  libraries like numpy, matplotlib, seaborn etc
import matplotlib.pyplot as plt
plt.plot (x.data(),sinx.data())
plt.plot(x.data(), cosx.data())
```
